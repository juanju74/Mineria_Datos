## 5. Limpieza de datos

### ¿Qué se hace?
Se identifican valores nulos y duplicados.

## 6. Análisis estadístico

### ¿Qué se hace?
Se calculan media, mediana, moda y dispersión.

### ¿Por qué?
Para entender el comportamiento general de los datos.

### ¿Qué significa?
- Media: promedio
- Mediana: valor central
- Moda: más frecuente
- Desviación: variabilidad


```python
col='CALIFICACIÓN FINAL CALIDAD'
df[col].mean(), df[col].median(), df[col].mode()
```

```python
df[col].std(), df[col].var()
```

## 7. Histograma

### ¿Qué se hace?
Se analiza la distribución de la calidad.

### ¿Por qué?
Para ver cómo se comportan los datos.

### ¿Qué se observa?
La mayoría de los datos se concentra entre valores altos.


```python
sns.histplot(df[col], kde=True)
plt.show()
```

## 8. Boxplot

### ¿Qué se hace?
Se identifican valores atípicos.

### ¿Por qué?
Para detectar datos extremos.

### ¿Qué se observa?
Existen datasets con baja calidad.


```python
sns.boxplot(x=df[col])
plt.show()
```

## 9. Categorías

### ¿Qué se hace?
Se analiza la distribución por categorías.

### ¿Por qué?
Para entender qué temas tienen más datos.

### ¿Qué se observa?
Algunas categorías dominan.


```python
sns.countplot(y=df['Categoría'])
plt.show()
```

## 10. Departamentos

### ¿Qué se hace?
Se analiza la distribución geográfica.

### ¿Por qué?
Para identificar diferencias regionales.

### ¿Qué se observa?
Existe desigualdad en la publicación de datos.


```python
sns.countplot(y=df['Departamento'])
plt.show()
```

## 11. Entidades

### ¿Qué se hace?
Se identifican las entidades que más publican.

### ¿Por qué?
Para analizar el nivel de participación institucional.

### ¿Qué se observa?
Algunas entidades lideran la publicación de datos.


```python
df['Nombre entidad'].value_counts().head(10).plot(kind='bar')
plt.show()
```

## 12. Heatmap

### ¿Qué se hace?
Se analiza la correlación entre variables numéricas.

### ¿Por qué?
Para identificar relaciones entre variables.

### ¿Qué se observa?
Existe una correlación moderada entre calidad y excelencia.


```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

## 13. Conclusión

### ¿Qué se concluye?
- Los datos tienen buena calidad en general
- Existen valores atípicos
- Hay desigualdad en la distribución
