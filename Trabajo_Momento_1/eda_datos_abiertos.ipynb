## 1. Importar librerías
Se importan las librerías necesarias para el análisis de datos y visualización.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## 2. Cargar el dataset
Se carga el archivo CSV que contiene la información de los datasets publicados.

```python
df = pd.read_csv('Vista_Reporte_de_Calidad_de_Datos_Abiertos_2025_20260315.csv')
df.head()
```

## 3. Dimensiones del dataset
Se analiza el número de filas y columnas.

```python
df.shape
```

## 4. Información del dataset
Se revisan tipos de datos y valores nulos.

```python
df.info()
```

## 5. Estadísticas descriptivas
Se calculan estadísticas básicas de las variables numéricas.

```python
df.describe()
```

## 6. Identificación de valores nulos
Se identifican los valores faltantes en cada columna.

```python
df.isnull().sum()
```

## 7. Limpieza de datos
Se reemplazan valores nulos en variables categóricas.

```python
df.fillna('No reportado', inplace=True)
```

## 8. Revisión de duplicados
Se verifica si existen registros duplicados.

```python
df.duplicated().sum()
```

## 9. Media, mediana y moda
Se calculan medidas de tendencia central para la variable de calidad.

```python
col = 'CALIFICACIÓN FINAL CALIDAD'

media = df[col].mean()
mediana = df[col].median()
moda = df[col].mode()[0]

media, mediana, moda
```

## 10. Medidas de dispersión
Se calcula la desviación estándar y varianza.

```python
std = df[col].std()
var = df[col].var()

std, var
```

## 11. Detección de outliers
Se utiliza un boxplot para identificar valores atípicos.

```python
sns.boxplot(x=df[col])
plt.title('Outliers en la calidad de datasets')
plt.show()
```

## 12. Histograma
Permite visualizar la distribución de la calidad.

```python
sns.histplot(df[col], bins=30)
plt.title('Distribución de calidad de datos')
plt.show()
```

## 13. Distribución por categoría
Se analiza la cantidad de datasets por categoría.

```python
plt.figure(figsize=(10,6))
sns.countplot(y=df['Categoría'])
plt.title('Datasets por categoría')
plt.show()
```

## 14. Distribución por departamento

```python
plt.figure(figsize=(10,8))
sns.countplot(y=df['Departamento'])
plt.title('Datasets por departamento')
plt.show()
```

- El análisis exploratorio permite entender la distribución y calidad de la información publicada.
